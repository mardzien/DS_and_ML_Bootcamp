```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r*')
```

```python
plt.plot(x, y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Titleee')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])

axes1.plot(x, y)
axes1.set_title("Bigger plot")
axes2.plot(y, x)
axes2.set_title("Smaller plot")
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2)

# axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First plot')
axes[1].plot(y, x)
axes[1].set_title('Second plot')
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,9))

ax[0].plot(x, y)
ax[1].plot(y, x)
```

```python
fig.savefig('my_pic.pdf')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, label='x squared')
ax.plot(x, x ** 3, label='x cubed')

ax.legend()
```
