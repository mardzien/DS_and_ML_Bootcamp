```python
# now emp_length is categorical column. We should map it to numeric. Where < 1 year = 0 and 10+ years = 10
```

```python
def map_emp_length(emp_length):
    if emp_length == '< 1 year':
        return 0
    elif emp_length == '10+ years':
        return 10
    else:
        return int(emp_length.split()[0])
```

```python
df['emp_length'] = df['emp_length'].apply(lambda x: map_emp_length(x) if isinstance(x, str) else x)
```

```python
# we have 7 unique values, A means the lowest risk, and G means the highest risk. I assume higher value would mean better grade
```

```python
def map_grade(grade):
    map_dict = {k: v for (v, k ) in enumerate('ABCDEFG'[::-1])}
    return map_dict[grade]
```

```python
df['grade'] = df['grade'].apply(map_grade)
```

```python
# we can create separate sub grade, so we need to remove letter from all values and leave digit only. Lower number means lower risks, so we need to map it to get reversed result
```

```python
def map_sub_grades(sub_grade):
    l = list(range(5))[::-1]
    return l[int(sub_grade) - 1]
```

```python
df['sub_grade'] = df['sub_grade'].apply(map_sub_grades)
```

```python
# let's map those values. Closer to ownership means bigger number
```

```python
def map_home_ownership(home_ownership):
    if home_ownership == 'OWN':
        return 3
    elif home_ownership == 'MORTGAGE':
        return 2
    elif home_ownership == 'RENT':
        return 1
    else:
        return 0
```

```python
from datetime import datetime

def map_short_month_to_number(date):
    month_part = date.split('-')[0]
    m_num = datetime.strptime(month_part, '%b').month
    return m_num
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('loan_status', axis=1).values
y = final_df['loan_status'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1)
```

```python
model = Sequential()

model.add(Dense(24, activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(12, activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=25, batch_size=48, validation_data=(X_test, y_test), callbacks=[es])
```

```python
from keras.models import load_model
```

```python
model.save('model.h5')
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = model.predict(X_test).round()
```

```python
print(classification_report(y_test, predictions))
```
